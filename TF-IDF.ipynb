```python
import math
import pandas as pd
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

def tf(d,fq):
    t={}
    for word in fq:
        t[word]=d.count(word)/len(d.split())
    return t

def idf(n,fq,d,docs):
    id={}
    for word in fq:
        s=0
        for doc in docs:
            if word in doc:
                   s+=1
        if word in d:
            id[word]=math.log(n/s)
        else:
            id[word]=0
    return id

def tf_idf(t,i):
    t_id={}
    for word in t:
        t_id[word]=t[word]*i[word]
    return t_id

n=int(input("Enter the number of documents:"))
docs=[]
text=""
for i in range(n):
    doc=input("Enter the document"+str(i+1)+": ")
    docs.append(doc.lower())
    text=text+doc.lower()+" "
tok=word_tokenize(text)
filtered_tok=[word.lower() for word in tok if word.isalpha()]
stopword_removed=[word for word in filtered_tok if word not in stopwords.words("english")]
fq=FreqDist(stopword_removed)
print("Cleaned data is:")
for i in fq:
    print("\t",i,":",fq[i])
tf_total=[]
for d in docs:
    tf_total.append(tf(d,fq))
idf_total=[]
for d in docs:
    idf_total.append(idf(n,fq,d,docs))
tf_idf_total=[]
for t,i in zip(tf_total,idf_total):
    tf_idf_total.append(tf_idf(t,i))
df=pd.DataFrame(tf_idf_total,index=["Doc"+str(i+1) for i in range(n)])
print("TF-IDF of each word in the given documents:")
df
```
